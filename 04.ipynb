第4章: 形態素解析
夏目漱石の小説『吾輩は猫である』の文章（neko.txt）をMeCabを使って形態素解析し，
その結果をneko.txt.mecabというファイルに保存せよ．このファイルを用いて，以下の問に対応するプログラムを実装せよ．

なお，問題37, 38, 39はmatplotlibもしくはGnuplotを用いるとよい．

```python
# MeCabのインポート
import MeCab

with open("neko.txt", 'r', encoding='utf-8') as f:
    s = f.read()
tagger = MeCab.Tagger ()

with open("neko.txt.mecab", 'w', encoding='utf-8') as f:
    f.write(tagger.parse(s).replace('\t',','))
```

```python
"""
30. 形態素解析結果の読み込み
形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．
ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとする
マッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．
第4章の残りの問題では，ここで作ったプログラムを活用せよ．
"""
sentens_list = []
with open("neko.txt.mecab", 'r', encoding='utf-8') as f:
    sentens = []
    for line in f:
        word = list(line.split(','))
        if len(word) > 1:
            pos = word[4].split('-')
            word_dic = {'surface':word[0],'base':word[3],'pos':pos,'pos1':word[5]}
            sentens.append(word_dic)
            if word[0] == '。':
                sentens_list.append(sentens)
                sentens = []
```

```python
"""
31. 動詞
動詞の表層形をすべて抽出せよ．
"""
verbs = []
for sentens in sentens_list:
    for word in sentens:
        if word['pos'][0] == "動詞":
            verbs.append(word['surface'])
print(verbs)
```

```python
"""
32. 動詞の基本形
動詞の基本形をすべて抽出せよ．
"""
bases = []
for sentens in sentens_list:
    for word in sentens:
        if word['pos'][0] == "動詞":
            bases.append(word['base'])
print(bases)
```

```python
"""
33. 「AのB」
2つの名詞が「の」で連結されている名詞句を抽出せよ．
"""
phrase = []
for sentens in sentens_list:
    prev1 = ""
    prev2 = ""
    for i in range(len(sentens)):
        if i == 0:
            prev2 = sentens[i]
            continue
        if i == 1:
            prev1 = sentens[i]
        if prev2['pos'][0] == "名詞" and prev1['surface'] == "の" and sentens[i]['pos'][0] == "名詞":
            phrase.append(prev2['surface']+prev1['surface']+sentens[i]['surface']) 
        prev2 = prev1
        prev1 = sentens[i]

print(phrase)        
```

```python
"""
34. 名詞の連接
名詞の連接（連続して出現する名詞）を最長一致で抽出せよ．
"""
phrases = []
cnt = 0
for sentens in sentens_list:
    phrase = ""
    for word in sentens:
        if word['pos'][0] == "名詞":
            phrase += word['surface']
            cnt += 1
        else:
            if cnt >= 2:
                phrases.append(phrase)
            phrase = ""
            cnt = 0
print(phrases)
            
```

```python
"""
35. 単語の出現頻度
文章中に出現する単語とその出現頻度を求め，出現頻度の高い順に並べよ
"""
surface_list = []
for sentens in sentens_list:
    for word in sentens:
        surface_list.append(word['surface'])

cnt_list = {}
for surface in set(surface_list):
    cnt_list[surface] = surface_list.count(surface)

sort_cnt_list = sorted(cnt_list.items(), key=lambda x: x[1], reverse=True)
print(sort_cnt_list[:10])


```

```python
"""
36. 頻度上位10語
出現頻度が高い10語とその出現頻度をグラフ（例えば棒グラフなど）で表示せよ．
"""

import matplotlib.pyplot as plt
import japanize_matplotlib

x = list(range(1, 11))
y = []
labels = []
for kye, value in sort_cnt_list[:10]:
    labels.append(kye)
    y.append(value)

plt.bar(x, y)
plt.xticks(x, labels)
plt.show()
```

```python
"""
37. 「猫」と共起頻度の高い上位10語
「猫」とよく共起する（共起頻度が高い）10語とその出現頻度をグラフ（例えば棒グラフなど）で表示せよ．
"""
import matplotlib.pyplot as plt
import japanize_matplotlib

words_with_cat = []
for sentens in sentens_list:
    words = []
    for word in sentens:
        words.append(word['surface'])
    if '猫' in words:
        words_with_cat.extend(words)

words_list = [word for word in words_with_cat if word not in '猫']

cnt_list_with_cat = {}
for word in set(words_list):
    cnt_list_with_cat[word] = words_list.count(word)

sort_cnt_list = sorted(cnt_list_with_cat.items(), key=lambda x: x[1], reverse=True)

x = list(range(1, 11))
y = []
labels = []
for kye, value in sort_cnt_list[:10]:
    labels.append(kye)
    y.append(value)
plt.bar(x, y)
plt.xticks(x, labels)
plt.show()
```

```python
"""
38. ヒストグラム
単語の出現頻度のヒストグラムを描け．
ただし，横軸は出現頻度を表し，1から単語の出現頻度の最大値までの線形目盛とする．
縦軸はx軸で示される出現頻度となった単語の異なり数（種類数）である．
"""
import matplotlib.pyplot as plt
import japanize_matplotlib
mode_list = []
for key, value in sort_cnt_list:
    mode_list.append(value)
    
plt.hist(mode_list,bins=100)
plt.show()
```

```python
"""
39. Zipfの法則
単語の出現頻度順位を横軸，その出現頻度を縦軸として，両対数グラフをプロットせよ．
"""
import matplotlib.pyplot as plt
import japanize_matplotlib
import numpy as np

x = []
y = []
for idx, mode in enumerate(sort_cnt_list):
    x.append(idx+1)
    y.append(mode[1])
    
plt.scatter(x,y)
plt.xscale('log')
plt.yscale('log')
plt.show()

    
```
